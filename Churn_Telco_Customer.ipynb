```python
df = df.drop_duplicates()
```

### **Data Manipulation**

```python
columns = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies','MultipleLines']
for column in columns:
    df[column] = df[column].replace({'No internet service':'No'})
    df[column] = df[column].replace({'No phone service':'No'})
```

```python
columns = ['gender', 'Partner', 'Dependents',
           'PhoneService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
           ,'Churn','MultipleLines']
for column in columns:
    df[column] = df[column].replace({'Yes':'1','No':'0'})
    df[column] = df[column].replace({'Male':'1','Female':'0'})
```

```python
df['TotalCharges'] = df.TotalCharges.replace(' ',-1)
df['TotalCharges'] = df.TotalCharges.astype(float)
ttcmean = df.TotalCharges.mean()
df['TotalCharges'] = df.TotalCharges.replace('none',ttcmean)
df
```

```python
df['InternetService_Fiber'] = (df['InternetService'] == 'Fiber optic').map({True: '1', False: '0'})
df['InternetService_DSL'] = (df['InternetService'] == 'DSL').map({True: '1', False: '0'})
df['InternetService_No'] = (df['InternetService'] == 'No').map({True: '1', False: '0'})
df.drop('InternetService', axis=1, inplace=True)

df['Month-to-month'] = (df['Contract'] == 'Month-to-month').map({True: '1', False: '0'})
df['1-Year'] = (df['Contract'] == 'One year').map({True: '1', False: '0'})
df['2-Year'] = (df['Contract'] == 'Two year').map({True: '1', False: '0'})
df.drop('Contract', axis=1, inplace=True)
df
```

# **Data Preprocessing**

```python
df = df.drop(columns = ["PaymentMethod","PaperlessBilling","MonthlyCharges"])
df.drop('customerID', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Churn', axis=1)
y = df.Churn

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

# **Train Model and Evaluate**

```python
# import the class
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

results = {}

# instantiate the model with a higher value for max_iter
logreg = LogisticRegression(max_iter=159)
# fit the model with data
logreg.fit(X_train, y_train)

y_test_pred = logreg.predict(X_test)
accuracy = accuracy_score(y_test, y_test_pred)
results['accuracy'] = accuracy

print(f"Accuracy: {accuracy}")
```

# **Re-Train Model**

```python
from imblearn.combine import SMOTEENN
sm = SMOTEENN()
X_res, y_res = sm.fit_resample(X, y)

Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_res, y_res, test_size=0.4, random_state=42)

logreg = LogisticRegression(max_iter=1000, class_weight='balanced')
logreg.fit(Xr_train, yr_train)

y_test_pred = logreg.predict(Xr_test)

accuracy = accuracy_score(yr_test, y_test_pred)
results['accuracy'] = accuracy

print(f"Accuracy: {accuracy}")
```
